```python
import seaborn as sns
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
import pandas as pd
df = pd.read_csv('data.csv')
df
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='Category',y ='Value',data= df)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='Product',y ='Sales',data= df)
```

```python
sns.scatterplot(x='total_bill' , y = 'tip' ,data = tips)
plt.title("Scatter plot of totalbill vs tips")
plt.show()
```

```python
# line plot
sns.lineplot(x='size',y='total_bill',data = tips)
```

```python
# categorical plot
# bar plot 

sns.barplot(x = "day", y = 'total_bill', data = tips)
plt.title("Bar Plot of TotalBill by Day")
plt.show()
```

```python
sns.catplot(x = "day", y = 'total_bill', hue='tip',data = tips)
plt.title("cat Plot of TotalBill by Day")
plt.show()
```

```python
sns.boxenplot(x = "day", y = 'total_bill', hue='tip',data = tips)
plt.title("boxen Plot of TotalBill by Day")
plt.show()
```

```python

sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
sns.violinplot(x='day', y = 'total_bill',data = tips)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde= True)
```

```python
sns.kdeplot(tips['total_bill'],fill = True)
```

```python
sns.pairplot(tips)
```

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y ='Total Revenue',data = sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y ='Total Revenue',data = sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
