```python
#using function add to list to create averge 3rd list
'''
l1=[10,20,30]
l2=[5,15,25]
l3=[(10+5)/2,(20+15)/2,(30+25)/2]
'''
def listadder(l1,l2):
    temp=[]
    for i in range(len(l1)):
        temp.append((l1[i]+l2[i])/2)
    return temp

l1=[10,20,30]
l2=[5,15,25]
l3=listadder(l1,l2)
print("average of 2 is:",l3)
```

```python
#find leader in given list:

#leade is on who has all prior smaller
#output:3,7,9,10 shout be in a list produced by
#function called LeaderGenerator()
'''
#output: list--->condition--->number
#process: ith is a leader iff all other from 0 to i-1 are lesser
          loop i for  ith number
            loop j to check from 0 to i-1
                if pass leader then add to leader list
#input: list to check on
'''
def LeaderGenerator(l):
    leader=[]
    for i in range(len(l)):
        flag=True
        for j in range(i):
            if l[j]>l[i]:
                flag=False
        if flag: #flag==True
            leader.append(l[i])
    return(leader)



l=[3,1,7,2,9,10]
leader_list=LeaderGenerator(l)
print("Leaders are:",leader_list)
```

```python
 '''Find Common Elements in Two Lists:
○ Question: Write a function to find common elements between two lists.
○ Logic: Use list comprehensions or a loop to compare elements.
○ Sample Input: [1, 2, 3, 4] and [3, 4, 5, 6]
○ Expected Output: [3, 4]
output: list of common elements in both
process: element to element comparission(in/not in)------use set
         def intersection(l1,l2)
input: 2 lists
'''
def intersection(l1,l2):
    common_list=[]
    for i in l1:
        if i in l2:
            common_list.append(i)
    return common_list
def intersection2(l1,l2):
    return list(set(l1)&set(l2))

l1=[1, 2, 3, 4]
l2=[3, 4, 5, 6]
common=intersection2(l1,l2)
print("common to both:",common)

```

```python
'''
Remove All Occurrences of an Element in a List:
Question: Write a function to remove all occurrences of a specific
element from a list.
Sample

Expected Output: List: [1, 3, 4]
process:count total time it is there,remove each time
                2--count 3---- call remove 3 times
Input: List: [1, 2, 2, 3, 4, 2] and Element: 2
'''
def remove_from_list(l,key):
    c=l.count(key)
    for i in range(c):
        l.remove(key)

l=[1, 2, 2, 3, 4, 2]
key=4
remove_from_list(l,key)
print("list after removal:",l)
```

```python
#Write a function to find the first non-repeating element in a list.
'''
start with each element stop first whos count is 1
Sample Input: [1, 2, 3, 2, 1, 4, 5]
output:3
'''
def non_repeating(data):
    for i in data:#read list element by element
        if data.count(i)==1:#if count is 1 return element and stop
            return i
data=[1, 2, 3, 2, 1, 4,3,1,5]
print("First non-repeating in list is:",non_repeating(data))
```
